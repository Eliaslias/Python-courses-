```python
#90. Filtering with String Methods
chicago = pd.read_csv("chicago.csv").dropna(how ="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
#So let's say I want to extract any rows from this data frame where the position title contains the word "Water
mask = chicago["Position Title"].str.lower().str.contains("water")
chicago[mask]
```

```python
#contains as I mentioned we'll look for that substring anywhere within the string.
#There's also a complimentary starts with an ends with strings method and those are only going to look
#at the beginning or end of the string.
```

```python
#i only want values or those rows that have water at the very beginning not just anywhere but only at the beginning.
chicago[chicago["Position Title"].str.lower().str.startswith("water")]
```

```python
#So let's extract all the rows where the value in the position title column ends with IST
mask1 = chicago["Position Title"].str.lower().str.endswith("ist")
chicago[mask1]
```

```python
#91. More String Methods - strip, lstrip, and rstrip
#And those are the strip strip and our strip methods and the L and L strip stands for left and the right
```

```python
chicago = pd.read_csv("chicago.csv").dropna(how ="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
# L strip method which is short for left strip and that removes any whitespace from the begining of the strip
"      hello world       ".lstrip()
```

```python
"      hello world       ".rstrip()
```

```python
#It removes whitespace from both the beginning and the end of a string.
"      hello world       ".strip()
```

```python
chicago["Name"] = chicago["Name"].str.rstrip().str.lstrip()
```

```python
chicago["Position Title"] =chicago["Position Title"].str.strip()
```

```python
## 95. The expand and n Parameters of the .str.split() Method
```

```python
chicago = pd.read_csv("chicago.csv").dropna(how ="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.head(3)
```

```python
chicago["Name"].str.split(", ", expand=True)
```

```python
chicago[["First Name","Last Name"]] = chicago["Name"].str.split(", ", expand=True)
```

```python
chicago.head(3)
```

```python
chicago["Position Title"].str.split(" ", expand=True, n =1)
```

```python
chicago[["First Title Word","Remaining word"]]= chicago["Position Title"].str.split(" ", expand=True, n =1)
```
