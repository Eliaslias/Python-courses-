```python
#155. Bar Graphs
#bar charts are really optimal for situations where you want to show things like counts especially of
#entities that are independent of each other.
google = data.DataReader(name="GOOG", data_source="yahoo", start= "2004-01-01", end= "2016-12-31")
google.head(3)
```

```python
def rank_performance(stock_price):
    if stock_price <=200:
        return "poor"
    elif stock_price > 200 and stock_price <= 500:
        return "Satifsfactory"
    else:
        return "Stellar"
```

```python
google["Close"].apply(rank_performance).value_counts()
```

```python
plt.style.use("ggplot")
google["Close"].apply(rank_performance).value_counts().plot(kind="line")
```

```python
plt.style.use("ggplot")
google["Close"].apply(rank_performance).value_counts().plot(kind="bar")
```

```python
plt.style.use("ggplot")
google["Close"].apply(rank_performance).value_counts().plot(kind="barh")
```

```python
# 156. Pie Charts 
#pie charts are really effective at showing how different parts come together to form the whole

apple = data.DataReader(name="AAPL", data_source="yahoo", start="2012-01-01", end= "2016-12-31")
apple.head(3)
```

```python
apple["Close"].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 93.3617192254726:
        return "Above Average"
    else:
        return "Below Average"
    
```

```python
apple["Close"].apply(rank_performance)
```

```python
apple["Close"].apply(rank_performance).value_counts()
```

```python
apple["Close"].apply(rank_performance).value_counts().plot(kind= "pie")
```

```python
apple["Close"].apply(rank_performance).value_counts().plot(kind= "pie",legend = True)
```
